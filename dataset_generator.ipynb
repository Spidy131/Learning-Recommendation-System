```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Parameters
num_users = 200
num_items = 150
num_records = 1000

user_ids = [f"U{str(i).zfill(4)}" for i in range(1, num_users+1)]
item_ids = [f"I{str(i).zfill(4)}" for i in range(1, num_items+1)]
item_types = ["course", "video", "quiz", "assignment", "reading"]
actions = ["viewed", "started", "completed", "attempted", "skipped"]
devices = ["web", "mobile"]
difficulty_levels = ["easy", "medium", "hard"]
topics = ["Python", "Math", "AI", "Robotics", "Data Science", "Electronics", "Statistics"]

# Updated structured topic mapping
learning_topics = [
    ("Python", "Basic", "Introduction to Python"),
    ("Python", "Basic", "Variables and Data Types"),
    ("Python", "Basic", "Control Flow and Loops"),

    ("Math", "Basic", "Number Systems and Arithmetic"),
    ("Math", "Basic", "Algebra Fundamentals"),
    ("Math", "Intermediate", "Linear Algebra"),

    ("AI", "Basic", "Introduction to AI Concepts"),
    ("AI", "Intermediate", "Machine Learning Algorithms"),
    ("AI", "Intermediate", "Neural Networks Basics"),

    ("Robotics", "Basic", "Introduction to Robotics"),
    ("Robotics", "Intermediate", "Kinematics and Dynamics"),
    ("Robotics", "Intermediate", "Control Systems"),

    # Newly added topics
    ("Electronics", "Basic", "Basic Circuit Components"),
    ("Electronics", "Intermediate", "Digital Electronics"),
    ("Electronics", "Intermediate", "Microcontrollers and Programming"),

    ("Statistics", "Basic", "Descriptive Statistics"),
    ("Statistics", "Intermediate", "Probability Distributions"),
    ("Statistics", "Intermediate", "Hypothesis Testing")
]

start_time = datetime(2025, 1, 1)

data = []

for _ in range(num_records):
    user = random.choice(user_ids)
    item = random.choice(item_ids)
    item_type = random.choice(item_types)
    action = random.choice(actions)
    rating = random.randint(1, 5) if action in ["completed", "attempted"] else None
    time_spent = random.randint(20, 2000) if action != "skipped" else 0
    device = random.choice(devices)
    difficulty = random.choice(difficulty_levels)
    tag = random.choice(topics)

    timestamp = start_time + timedelta(seconds=random.randint(0, 30000000))

    # Assign structured learning topic
    current_topic = random.choice(learning_topics)

    data.append([
        user, item, item_type, action, rating, time_spent, device,
        difficulty, tag, timestamp, current_topic[0], current_topic[1], current_topic[2]
    ])

df = pd.DataFrame(data, columns=[
    "user_id", "item_id", "item_type", "action",
    "rating_score", "time_spent_seconds", "device",
    "difficulty_level", "tags", "timestamp",
    "course_category", "course_level", "course_topic"
])

df.to_csv("learning_interactions_dataset_updated.csv", index=False)
print("Dataset generated: learning_interactions_dataset_updated.csv")
```
